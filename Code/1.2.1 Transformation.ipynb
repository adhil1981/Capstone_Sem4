```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming you have already read your dataset into a DataFrame named 'data'

# Distribution of Incidents per Day
data['Date'] = data['Dates'].dt.date

plt.figure(figsize=(10, 6))
data_per_day = data.groupby('Date').size()
sns.kdeplot(data=data_per_day, shade=True, color='pink')
plt.axvline(x=data_per_day.median(), ymax=0.95, linestyle='--', color='purple')
plt.annotate(
    'Median: ' + str(data_per_day.median()),
    xy=(data_per_day.median(), 0.004),
    xytext=(200, 0.005),
    arrowprops=dict(arrowstyle='->', color='purple', shrinkB=10))
plt.title('Distribution of Incidents per Day', fontdict={'fontsize': 16})
plt.xlabel('Incidents')
plt.ylabel('Density')
plt.legend().remove()
plt.show()

# Distribution of Incidents per Hour
data['Hour'] = data['Dates'].dt.hour

plt.figure(figsize=(10, 6))
data_per_hour = data.groupby('Hour').size()
sns.kdeplot(data=data_per_hour, shade=True, color='pink')
plt.axvline(x=data_per_hour.median(), ymax=0.95, linestyle='--', color='purple')
plt.annotate(
    'Median: ' + str(data_per_hour.median()),
    xy=(data_per_hour.median(), 0.025),
    xytext=(5, 0.03),
    arrowprops=dict(arrowstyle='->', color='purple', shrinkB=10))
plt.title('Distribution of Incidents per Hour', fontdict={'fontsize': 16})
plt.xlabel('Hour')
plt.ylabel('Density')
plt.legend().remove()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming you have already read your dataset into a DataFrame named 'data'

# Extract the date from the 'Dates' column
data['Date'] = data['Dates'].dt.date

# Group the data by 'Hour', 'Date', and 'Category', count incidents, and keep only relevant columns
data = data.groupby(['Hour', 'Date', 'Category'], as_index=False).count().iloc[:, :4]
data.rename(columns={'Dates': 'Incidents'}, inplace=True)

# Group the data by 'Hour' and 'Category', calculate the mean of incidents
data = data.groupby(['Hour', 'Category'], as_index=False).mean()

# Filter data to include only specified crime categories
data = data.loc[data['Category'].isin(
    ['OTHER OFFENCE', 'THEFT', 'ASSAULT', 'FRAUD', 'DRUNKENNESS', 'KIDNAPPING', 'PROSTITUTION'])]

# Plot the average number of incidents per hour for each crime category
sns.set_style("whitegrid")
fig, ax = plt.subplots(figsize=(14, 4))
ax = sns.lineplot(x='Hour', y='Incidents', data=data, hue='Category')
ax.legend(loc='upper center', bbox_to_anchor=(0.5, 1.15), ncol=6)
plt.suptitle('Average number of incidents per hour')
fig.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python
# my personal reusable function for detecting missing data
def missing_value_describe(data):
    # check missing values in training data
    missing_value_stats = (data.isnull().sum() / len(data)*100)
    missing_value_col_count = sum(missing_value_stats > 0)
    missing_value_stats = missing_value_stats.sort_values(ascending=False)[:missing_value_col_count]
    print("Number of columns with missing values:", missing_value_col_count)
    if missing_value_col_count != 0:
        # print out column names with missing value percentage
        print("\nMissing percentage (desceding):")
        print(missing_value_stats)
    else:
        print("No missing data!!!")
missing_value_describe(data)
```
