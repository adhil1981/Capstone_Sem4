```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming you have already read your dataset into a DataFrame named 'data'

# Convert the 'Dates' column to a pandas datetime object
data['Dates'] = pd.to_datetime(data['Dates'])

# Extract the hour from the 'Dates' column to determine day or night
data['Hour'] = data['Dates'].dt.hour

# Define a function to categorize each hour as day or night
def is_day_or_night(hour):
    if 6 <= hour < 18:
        return 'Day'
    else:
        return 'Night'

# Apply the function to create a new column 'Day_Night'
data['Day_Night'] = data['Hour'].apply(is_day_or_night)

# Get the top 10 police districts based on the number of crimes
top_10_districts = data['PdDistrict'].value_counts().nlargest(10)

# Filter the original DataFrame to include only the rows with the top 10 police districts
data_top_10 = data[data['PdDistrict'].isin(top_10_districts.index)]

# Reorder the police districts based on crime count (highest to lowest)
ordered_districts = top_10_districts.index

# Create a pivot table to calculate day and night crime counts for each district
crime_counts_pivot = data_top_10.pivot_table(index='PdDistrict', columns='Day_Night', aggfunc='size', fill_value=0)

# Bar Chart - Crime Distribution by Top 10 Police Districts (Highest to Lowest)
plt.figure(figsize=(10, 6))
sns.barplot(x=crime_counts_pivot.index, y=crime_counts_pivot['Day'], order=ordered_districts, color='pink', label='Day')
sns.barplot(x=crime_counts_pivot.index, y=crime_counts_pivot['Night'], order=ordered_districts, color='purple', label='Night')
plt.title('Top 10 Police Districts by Day and Night Crime Count (Highest to Lowest)')
plt.xlabel('Police District')
plt.ylabel('Number of Crimes')
plt.legend()

# Add day and night crime counts as labels above each bar
for index, district in enumerate(ordered_districts):
    day_count = crime_counts_pivot.loc[district, 'Day']
    night_count = crime_counts_pivot.loc[district, 'Night']
    plt.text(index, day_count, str(day_count), ha='center', va='bottom', fontsize=10, color='black')
    plt.text(index, day_count + night_count, str(night_count), ha='center', va='bottom', fontsize=10, color='black')

# Remove the grid lines
plt.grid(False)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming you have already read your dataset into a DataFrame named 'data'

# Convert the 'Dates' column to a pandas datetime object
data['Dates'] = pd.to_datetime(data['Dates'])

# Extract the hour from the 'Dates' column to determine day or night
data['Hour'] = data['Dates'].dt.hour

# Define a function to categorize each hour as day or night
def is_day_or_night(hour):
    if 6 <= hour < 18:
        return 'Day'
    else:
        return 'Night'

# Apply the function to create a new column 'IsDay'
data['IsDay'] = data['Hour'].apply(is_day_or_night)

# Create a cross-tabulation of crime categories by day/night
ct = pd.crosstab(data['Category'], data['IsDay'])

# Plot the horizontal bar chart
plt.figure(figsize=(14, 14))
ct.plot(kind='barh', ax=plt.gca(), color=['skyblue', 'purple'])
plt.title('Crime Categories by Day/Night')
plt.xlabel('Number of Crimes')
plt.ylabel('Crime Category')

# Add crime counts as labels inside each bar
for i in range(ct.shape[0]):
    day_count = ct.iloc[i, 0]
    night_count = ct.iloc[i, 1]
    plt.text(day_count / 2, i, f'{day_count:,}', ha='center', va='center', fontsize=10, color='black')
    plt.text(day_count + night_count / 2, i, f'{night_count:,}', ha='center', va='center', fontsize=10, color='black')

# Show the plot
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming you have already read your dataset into a DataFrame named 'data'

# Convert the 'Dates' column to a pandas datetime object
data['Dates'] = pd.to_datetime(data['Dates'])

# Filter the data for the years 2018 to 2023
data_2018_to_2023 = data[data['Dates'].dt.year.between(2018, 2023)]

# Get the top 7 crime categories for each year
top_7_crime_categories = data_2018_to_2023.groupby([data_2018_to_2023['Dates'].dt.year, 'Category']).size().reset_index(name='Count')
top_7_crime_categories = top_7_crime_categories.groupby('Dates').apply(lambda x: x.nlargest(7, 'Count')).reset_index(drop=True)

# Create a grouped bar chart
plt.figure(figsize=(12, 6))
sns.barplot(x='Dates', y='Count', hue='Category', data=top_7_crime_categories, palette='viridis')
plt.title('Top 7 Crime Categories Year-wise (2018 to 2023)')
plt.xlabel('Year')
plt.ylabel('Number of Crimes')
plt.legend(title='Crime Category', bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=8)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming you have already read your dataset into a DataFrame named 'data'

# Distribution of Incidents per Day
data['Date'] = data['Dates'].dt.date

plt.figure(figsize=(10, 6))
data_per_day = data.groupby('Date').size()
sns.kdeplot(data=data_per_day, shade=True, color='pink')
plt.axvline(x=data_per_day.median(), ymax=0.95, linestyle='--', color='purple')
plt.annotate(
    'Median: ' + str(data_per_day.median()),
    xy=(data_per_day.median(), 0.004),
    xytext=(200, 0.005),
    arrowprops=dict(arrowstyle='->', color='purple', shrinkB=10))
plt.title('Distribution of Incidents per Day', fontdict={'fontsize': 16})
plt.xlabel('Incidents')
plt.ylabel('Density')
plt.legend().remove()
plt.show()

# Distribution of Incidents per Hour
data['Hour'] = data['Dates'].dt.hour

plt.figure(figsize=(10, 6))
data_per_hour = data.groupby('Hour').size()
sns.kdeplot(data=data_per_hour, shade=True, color='pink')
plt.axvline(x=data_per_hour.median(), ymax=0.95, linestyle='--', color='purple')
plt.annotate(
    'Median: ' + str(data_per_hour.median()),
    xy=(data_per_hour.median(), 0.025),
    xytext=(5, 0.03),
    arrowprops=dict(arrowstyle='->', color='purple', shrinkB=10))
plt.title('Distribution of Incidents per Hour', fontdict={'fontsize': 16})
plt.xlabel('Hour')
plt.ylabel('Density')
plt.legend().remove()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming you have already read your dataset into a DataFrame named 'data'

# Extract the date from the 'Dates' column
data['Date'] = data['Dates'].dt.date

# Group the data by 'Hour', 'Date', and 'Category', count incidents, and keep only relevant columns
data = data.groupby(['Hour', 'Date', 'Category'], as_index=False).count().iloc[:, :4]
data.rename(columns={'Dates': 'Incidents'}, inplace=True)

# Group the data by 'Hour' and 'Category', calculate the mean of incidents
data = data.groupby(['Hour', 'Category'], as_index=False).mean()

# Filter data to include only specified crime categories
data = data.loc[data['Category'].isin(
    ['OTHER OFFENCE', 'THEFT', 'ASSAULT', 'FRAUD', 'DRUNKENNESS', 'KIDNAPPING', 'PROSTITUTION'])]

# Plot the average number of incidents per hour for each crime category
sns.set_style("whitegrid")
fig, ax = plt.subplots(figsize=(14, 4))
ax = sns.lineplot(x='Hour', y='Incidents', data=data, hue='Category')
ax.legend(loc='upper center', bbox_to_anchor=(0.5, 1.15), ncol=6)
plt.suptitle('Average number of incidents per hour')
fig.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python
# my personal reusable function for detecting missing data
def missing_value_describe(data):
    # check missing values in training data
    missing_value_stats = (data.isnull().sum() / len(data)*100)
    missing_value_col_count = sum(missing_value_stats > 0)
    missing_value_stats = missing_value_stats.sort_values(ascending=False)[:missing_value_col_count]
    print("Number of columns with missing values:", missing_value_col_count)
    if missing_value_col_count != 0:
        # print out column names with missing value percentage
        print("\nMissing percentage (desceding):")
        print(missing_value_stats)
    else:
        print("No missing data!!!")
missing_value_describe(data)
```
